```python
#problem 3 
import json 
import os
import stat
from pathlib import Path

data = {"name" : "Alice" , "age" : 25 , "city" : "New York"}
file_name = 'lab.json'
os.chmod(file_name , stat.S_IWRITE) #make the file read_write

def save (data , file_name) : 
    try : 
        path = Path(file_name)
        contents = json.dumps(data)
        path.write_text(contents)
    except IOError : 
        print(f"IOError while saving data to {file_name}")
    
    else : 
        print(f"Data Saved Successfully to {file_name}!!")
        
def load (file_name) : 
    try : 
        path = Path(file_name)
        contents = path.read_text()
        data = json.loads(contents)
    
    except IOError : 
        print(f"IOError while loading data from {file_name}")
        
    else : 
        return data 


save(data , file_name) 

retrieved_data = load(file_name)

print(f"Contents of {file_name} : ")
print(retrieved_data)


    
```

```python
from pathlib import Path
import json 
import os
import stat
data = {"name" : "Alice" , "age" : 25 , "city" : "New York"}

def main_menu_display () : 
    print ("1. Read a file\n2. Write to a file\n3. Save data to JSON\n4. Exit\n")

def get_option():
    while True:
            menu_option = int(input("Choose an option: "))
            if menu_option >=1 and menu_option <= 4:  
                return menu_option
            else:
                print("Please choose a valid option between 1 and 4.")
       
            
def process (num) : 
    if num == 1 : 
        read_file()
    elif num == 2 : 
        write_to_file()
    elif num == 3 :
        save_to_json(data)
    elif num == 4 : 
        exit()
        
def read_file() : 
    
    while True : 
        
        file_name = input("Enter file name : ")
        try : 
            path = Path(file_name)
            contents = path.read_text()

        except FileNotFoundError : 
            
            print(f"File {file_name} not found.")
            again_quit = input("Would you like to try again? (yes/no): ")
            
            if again_quit == "yes" : 
                continue 
            elif again_quit == "no" : 
                break
    
        else : 
            print(contents)
            break 
        
def write_to_file() :
    
    file_name = input("Enter file name : ")
#os.chmod(file_name , stat.S_IREAD)
    text_entered = input("Enter text to save: ")
    try : 
        path = Path(file_name)
        path.write_text(text_entered)
        contents = path.read_text()

    except PermissionError : 
        print(f"Permission denied to write to {file_name}")

    else : 
        print(f"Data saved successfully to {file_name}")
        print (f"Contents of {file_name}: ")
        print(contents)
    
def save_to_json (data) : 
    
    file_name = input ("Type in the json file: ")
    try : 
        path = Path(file_name)
        contents = json.dumps(data)
        path.write_text(contents)
        
    except IOError : 
        print(f"IOError while saving data to {file_name}")
    
    else : 
        print(f"Data Saved Successfully to {file_name}!!")
def exit() : 
    print("Goodbye!")
    
# main program 
main_menu_display ()
menu_option = get_option()

process(menu_option)
```

```python
import re 

while True : 
    name = input("Enter your name: ")
    if name.replace(" ","").isalpha() : 
        break 
    else : 
        print("Name must only contains alphabetic chars and spaces") 
        continue 


while True :
    try : 
        age = int(input("Enter your age: "))
        if age > 0 : 
            break 
        
        elif age < 0 : 
            print("Age must be postitive!")
            continue 
        
    except ValueError : 
        print("Age must be an integer!")
       
while True : 
    email = input("Enter your email: ")
    if re.match(r"[^@]+@[^@]+\.[^@]+", email):
        break 
    else : 
        print("Email must follow a standard email format!")

print("Validated Data: ")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Email: {email}")

```

```python
#Bonus problem

import json 
import os
import stat
from pathlib import Path

def get_name() : 
    while True : 
        name = input("Name: ")
        #isalpha() to ensure no characters inserted other than alphabets (rejects numbers and special chars)
        if isinstance(name , str) and name.isalpha() : 
            return name
        else :
            print("Name must be a string! try again.")
            continue
    
def get_age() :  
    while True : 
        try : 
            age = int(input("Age: "))
            if age <= 0 : 
                print("age must be positive. try again")
                continue
            else : 
                return age
        except ValueError : 
                print("age must be an integer. try again")
                continue
         
def get_city() : 
    while True : 
        city = input("city: ")
        if city.replace(" ","").strip() : 
            return city 
        else : 
            print("City must be non-empty string")
            continue          

        

file_name = 'lab.json'
name = get_name()
age = get_age()
city = get_city()
data = {"name" : name , "age" : int(age) , "city" : city} 
os.chmod(file_name , stat.S_IWRITE)

def save (data , file_name) : 
    try : 
        path = Path(file_name)
        contents = json.dumps(data)
        path.write_text(contents)
    except IOError : 
        print(f"IOError while saving data to {file_name}")
    
    else : 
        print(f"Data Saved Successfully to {file_name}!!")
        
def load (file_name) : 
    try : 
        path = Path(file_name)
        contents = path.read_text()
        data = json.loads(contents)
    
    except IOError : 
        print(f"IOError while loading data from {file_name}")
        
    else : 
        return data 


save(data , file_name) 

retrieved_data = load(file_name)

print(f"Contents of {file_name} : ")
print(retrieved_data)

```
