```python
#problem 1
import math
class Pagination : 
    """A class that handles paginated content in a website"""

    def __init__(self, items, page_size):
        self.items = items
        self.page_size = int(page_size)
        self.current_page = 1 
        self.total_pages = math.ceil(len(items)/page_size)
        
    def get_visible_items(self): 
        start = (self.current_page - 1) * self.page_size
        end = start + self.page_size
        print(self.items[start:end])
    
    def next_page (self): 
        if self.current_page >= self.total_pages : #if the current page is the last page >> the next will be the first page again (rotates forward) 
            self.current_page = 1
        else : 
            self.current_page += 1 
        
    def prev_page (self):
        if self.current_page <= 1 :
            self.current_page = self.total_pages #if the current page is the first page then the previous page will be the last one (rotates backward)
        else : 
            self.current_page -= 1
            
    def first_page (self) : 
        self.current_page = 1
        
    def last_page (self) : 
        self.current_page = self.total_pages
        
    def go_to_page (self , page ) :
        if page > self.total_pages :
            self.current_page = self.total_pages 
        elif page < 1 : 
            self.current_page = 1 
        else : 
            self.current_page = page
        
# i let chatgpt generate the list data for me (Honestly there's no time to think about the data I'm supposed to enter lol)
jobs = [
    "Software Developer",
    "Data Scientist",
    "Graphic Designer",
    "Civil Engineer",
    "Mechanical Engineer",
    "Project Manager",
    "Marketing Specialist",
    "Sales Executive",
    "Human Resources Manager",
    "Financial Analyst",
    "Accountant",
    "UX/UI Designer",
    "Cybersecurity Analyst",
    "Machine Learning Engineer",
    "Product Manager",
    "Customer Support Representative",
    "Content Writer",
    "Journalist",
    "Digital Marketing Manager",
    "Electrician",
    "Plumber",
    "Carpenter",
    "Doctor",
    "Nurse",
    "Pharmacist",
    "Veterinarian",
    "Teacher",
    "Professor",
    "Lawyer",
    "Paralegal",
    "Architect",
    "Interior Designer",
    "Chef",
    "Bartender",
    "Flight Attendant",
    "Pilot",
    "Real Estate Agent",
    "Social Media Manager",
    "Fitness Trainer",
    "Event Planner",
    "Photographer",
    "Videographer",
    "Software Tester",
    "IT Support Specialist",
    "Web Developer",
    "Cloud Engineer",
    "AI Researcher",
    "Business Analyst",
    "Environmental Scientist",
    "Biotechnologist",
    "Chemical Engineer",
    "Research Scientist"
]


p = Pagination(jobs, 4)

p.last_page()
p.get_visible_items()

p.next_page()
p.get_visible_items()

p.next_page()
p.get_visible_items()

p.prev_page()
p.get_visible_items()

p.prev_page()
p.get_visible_items()

p.first_page()
p.get_visible_items()
```

```python
class CoffeeShop : 
    
    def __init__(self , name , menu):
        self.name = name
        self.menu = menu 
        self.orders = []
        

    def add_order (self , order) :
        order_available = False #flag to track if is the order is available (to prevent executing both the append and print if the item exist)
        for item in self.menu : #iterate over each element (dictionary) in the list and then check if the added order matches the value of the key "name"
            if item["name"] == order : 
                self.orders.append(order)
                order_available = True
                break
        if order_available == False : 
            print(f"{order} is currently unavailable!")
            
    def fullfil_order (self) : 
        if self.orders == [] : 
            print("All orders have been fulfilled!")
        else : 
            print(f"The {self.orders[0]} is ready!")
            self.orders.pop(0)
     
    def list_orders(self) : 
        print(self.orders)
        
    def due_amount (self) : 
        total_due = 0
        for item in self.menu : # This iterates over the menu so that if any of its items exist in the orders list, the item's price is added
            if item["name"] in self.orders :
                total_due += item["price"]
        print(f"Due amount : {total_due}")

    def cheapest_item (self) : 
        cheapest_item = self.menu[0].copy() #make a copy of the first dictionary to compare with other dictionaries in the list
        for item in self.menu : 
            if cheapest_item["price"] > item["price"] : 
                cheapest_item = item.copy() #if the current cheapest item is more expensive than an item then assign it as the new current cheapest
        
        print(f"{cheapest_item["name"]} is the cheapest item on the menu - price : {cheapest_item["price"]}")
        
    def drink_only (self) : 
        drinks_list = []
        
        for item in self.menu : 
            if item["type"] == "drink" : 
                drinks_list.append(item["name"])
        
        print(drinks_list)
        
    def foods_only (self) : 
        foods_list = []
        
        for item in self.menu : 
            if item["type"] == "food" : 
                foods_list.append(item["name"])
        
        print(foods_list)


# i let chatgpt generate the menu list of dictionaries data (It'll be a headache to enter
# it myself, and i'll also be able to test the code better because it now has more and better data)    
  
menu = [
    {"name": "Burger", "type": "food", "price": 5.99},
    {"name": "Pizza", "type": "food", "price": 8.99},
    {"name": "Pasta", "type": "food", "price": 7.49},
    {"name": "Salad", "type": "food", "price": 4.99},
    {"name": "Sushi", "type": "food", "price": 12.99},
    {"name": "Steak", "type": "food", "price": 15.99},
    {"name": "Taco", "type": "food", "price": 3.49},
    {"name": "Fried Chicken", "type": "food", "price": 6.99},
    {"name": "French Fries", "type": "food", "price": 2.99},
    {"name": "Ice Cream", "type": "food", "price": 3.99},
    {"name": "Coffee", "type": "drink", "price": 2.49},
    {"name": "Tea", "type": "drink", "price": 1.99},
    {"name": "Soda", "type": "drink", "price": 1.49},
    {"name": "Juice", "type": "drink", "price": 2.99},
    {"name": "Milkshake", "type": "drink", "price": 4.49},
    {"name": "Water", "type": "drink", "price": 0.99},
    {"name": "Beer", "type": "drink", "price": 3.49},
    {"name": "Wine", "type": "drink", "price": 5.99},
    {"name": "Cocktail", "type": "drink", "price": 7.99},
    {"name": "Smoothie", "type": "drink", "price": 4.99}
]


cafe = CoffeeShop("Villa", menu)
cafe.foods_only()
cafe.drink_only()
cafe.cheapest_item()
cafe.due_amount()
cafe.add_order("Ice Cream")
cafe.list_orders()
cafe.due_amount()
cafe.add_order("Steak")
cafe.list_orders()
cafe.due_amount()
cafe.add_order("Sushi")
cafe.list_orders()
cafe.due_amount()
cafe.fullfil_order()
cafe.fullfil_order()
cafe.due_amount()
cafe.fullfil_order()
cafe.fullfil_order()
cafe.due_amount()
```
